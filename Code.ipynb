```python

import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from statistics import mean
from sklearn.inspection import permutation_importance
from matplotlib import pyplot

from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.model_selection import KFold, cross_val_score

dataset1=pd.read_csv('at1.csv')
dataset2=pd.read_csv('at2.csv')
dataset4=pd.read_csv('at4.csv')
dataset8=pd.read_csv('at8.csv')
dataset16=pd.read_csv('at16.csv')
datasetmodified16=pd.read_csv('modified-at16.csv')


def dataset_twoBSM(dataset,n):
    dataset=dataset.sample(frac = 1).reset_index(drop=True) 
    output=dataset[['pos-x1','pos-y1','spd-x1','spd-y1','pos-x2','pos-y2','spd-x2','spd-y2','AttackerType']]
    output1=output.drop_duplicates()
    X=output1[['pos-x1','pos-y1','spd-x1','spd-y1','pos-x2','pos-y2','spd-x2','spd-y2']]
    Y=output1['AttackerType']/n
    return X, Y
 
precision_mean1=[]
recall_mean1=[]
accuracy_mean1=[]
F1_mean1=[]
                     
precision_mean2=[]
recall_mean2=[]
accuracy_mean2=[]
F1_mean2=[]
                     
precision_mean4=[]
recall_mean4=[]
accuracy_mean4=[]
F1_mean4=[]
                     
precision_mean8=[]
recall_mean8=[]
accuracy_mean8=[]
F1_mean8=[]

precision_mean16=[]
recall_mean16=[]
accuracy_mean16=[]
F1_mean16=[]

precision_mean_mod16=[]
recall_mean_mod16=[]
accuracy_mean_mod16=[]
F1_mean_mod16=[]
                     

def datasetkfold(X, Y):
    kf = KFold(n_splits=5, shuffle=True)
    precision_mean=[]
    recall_mean=[]
    accuracy_mean=[]
    F1_mean=[]
    for train_index, test_index in kf.split(X):
        Xval_train, Xval_test = X.iloc[train_index], X.iloc[test_index]
        Yval_train, Yval_test = Y.iloc[train_index], Y.iloc[test_index]
        precision, recall, accuracy, F1=score(Xval_train, Yval_train, Xval_test, Yval_test)
        precision_mean.append(precision)
        recall_mean.append(recall)
        accuracy_mean.append(accuracy)
        F1_mean.append(F1)
        
    return precision_mean, recall_mean, accuracy_mean, F1_mean


def score(xtrain, ytrain, xtest, ytest):
    knn = KNeighborsClassifier(n_neighbors=3, weights='distance')
    #knn= RandomForestClassifier(n_estimators=5)  
    #knn=DecisionTreeClassifier()
    #knn=GaussianNB()
    knn=knn.fit(xtrain, ytrain)
    pred1 = knn.predict(xtest) 
    print("\nConfusion matrix:\n ",confusion_matrix(ytest, pred1))
    tn1, fp1, fn1, tp1 = confusion_matrix(ytest, pred1).ravel()
    precision= tp1/(tp1+fp1)
    recall=tp1/(tp1+fn1)
    accuracy=(tp1+tn1)/(tp1+tn1+fn1+fp1)
    F1=2*((precision*recall)/(precision+recall))
    
    #print(f" Precision: {precision*100}, Recall: {recall*100} and Accuracy: {accuracy*100} and F1: {F1*100}")

    #print("\nReport: ",classification_report(ytest, pred1))
          
    return precision*100, recall*100, accuracy*100, F1*100


X1, Y1=dataset_twoBSM(dataset1,1)
X2, Y2=dataset_twoBSM(dataset2,2)
X4, Y4=dataset_twoBSM(dataset4,4)   
X8, Y8=dataset_twoBSM(dataset8,8)
X16, Y16=dataset_twoBSM(dataset16,16)
Xm16, Ym16=dataset_twoBSM(datasetmodified16,16)


#For datasetkfold
print("ATTACK1 \n")
precision_mean1, recall_mean1, accuracy_mean1, F1_mean1= datasetkfold(X1,Y1)
print("ATTACK2 \n")
precision_mean2, recall_mean2, accuracy_mean2, F1_mean2= datasetkfold(X2,Y2)
print("ATTACK4 \n")
precision_mean4, recall_mean4, accuracy_mean4, F1_mean4= datasetkfold(X4,Y4)
print("ATTACK8 \n")
precision_mean8, recall_mean8, accuracy_mean8, F1_mean8= datasetkfold(X8,Y8)
print("ATTACK16 \n")
precision_mean16, recall_mean16, accuracy_mean16, F1_mean16= datasetkfold(X16,Y16)
print("Modified ATTACK16 \n")
precision_mean_mod16, recall_mean_mod16, accuracy_mean_mod16, F1_mean_mod16= datasetkfold(Xm16,Ym16)
print("\n\n")


print("\n ATTACK 1 \n")
print(f"Mean Precision: {mean(precision_mean1)},\n Mean Recall: {mean(recall_mean1)}, \n Mean Accuracy:{mean(accuracy_mean1)},\n Mean F1:{mean(F1_mean1)}")

print("\n ATTACK 2 \n")
print(f" Mean Precision: {mean(precision_mean2)},\n Mean Recall: {mean(recall_mean2)},\n Mean Accuracy:{mean(accuracy_mean2)},\n Mean F1:{mean(F1_mean2)}")

print("\n ATTACK 4 \n")
print(f"Mean Precision: {mean(precision_mean4)},\n Mean Recall: {mean(recall_mean4)}, \n Mean Accuracy:{mean(accuracy_mean4)},\n Mean F1:{mean(F1_mean4)}")

print("\n ATTACK 8 \n")
print(f" Mean Precision: {mean(precision_mean8)},\n Mean Recall: {mean(recall_mean8)},\n Mean Accuracy:{mean(accuracy_mean8)},\n Mean F1:{mean(F1_mean8)}")

print("\n ATTACK 16 \n")
print(f"Mean Precision: {mean(precision_mean16)},\n Mean Recall: {mean(recall_mean16)},\n Mean Accuracy:{mean(accuracy_mean16)},\n Mean F1:{mean(F1_mean16)}")

print("\n ATTACK Modified 16 \n")
print(f" Mean Precision: {mean(precision_mean_mod16)},\n Mean Recall: {mean(recall_mean_mod16)},\n Mean Accuracy:{mean(accuracy_mean_mod16)},\n Mean F1:{mean(F1_mean_mod16)}")


```
